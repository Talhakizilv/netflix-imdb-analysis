```python
import pandas as pd

df = pd.read_csv("NetflixOriginals.csv", encoding="cp1252")
```

```python
print("Rows,Columns:", df.shape)
print("\nColumns:", df.columns.tolist())

print("\nFirst 5 Row:")
print(df.head())
```

```python
df.columns = [c.strip() for c in df.columns]
```

```python
df["IMDB Score"] = pd.to_numeric(df["IMDB Score"], errors="coerce")
df["Runtime"] = pd.to_numeric(df["Runtime"], errors="coerce")

df["Premiere"] = pd.to_datetime(df["Premiere"], errors="coerce")
df["Year"] = df["Premiere"].dt.year
```

```python
df["Title"] = df["Title"].astype(str).str.strip()
df["Genre"] = df["Genre"].astype(str).str.strip()
df["Language"] = df["Language"].astype(str).str.strip()

before = df.shape[0]

df = df.dropna(subset=["IMDB Score", "Year", "Runtime", "Title", "Genre"])

after = df.shape[0]
print("The number of rows deleted:", before - after)

```

```python
df = df[(df["IMDB Score"] >= 0) & (df["IMDB Score"] <= 10)]
df = df[df["Runtime"] > 0]

dup_count = df.duplicated(subset=["Title"]).sum()

df["Main Genre"] = df["Genre"].str.split(",").str[0].str.strip()

df.to_csv("netflix_originals_cleaned.csv", index=False)



```
